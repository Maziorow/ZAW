```python
import matplotlib.pyplot as plt
```

## Visualize Event Data

```python
time = []
x = []
y = []
polarity = []

with open("shapes_rotation/events.txt", "r") as f:
    for line in f:
        parts = line.split(' ')
        if float(parts[0])>1:
            break
        time.append(float(parts[0]))
        x.append(float(parts[1]))
        y.append(float(parts[2]))
        polarity.append(float(parts[3]))

print(f'Amount of events: {len(time)}')
print(f'First timestamp: {time[0]}')
print(f'Last timestamp: {time[-1]}')
print(f'X min: {min(x)} and X max: {max(x)}')
print(f'Y min: {min(y)} and Y max: {max(y)}')
print(f'Positive events: {int(sum(polarity))}')
print(f'Negative events: {int(len(polarity)-sum(polarity))}')

colors = ['green' if p == 1 else 'red' for p in polarity]

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, time, c=colors, s=1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Time')
ax.set_title("Event camera")
plt.show()
```
