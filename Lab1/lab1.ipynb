## Zadanie 1.5

```python
I_mandril = cv2.imread("mandril.png")
I_lena = cv2.imread("lena.png")

I_mandril = cv2.cvtColor(I_mandril,cv2.COLOR_BGR2GRAY)
I_lena = cv2.cvtColor(I_lena,cv2.COLOR_BGR2GRAY)

fig, axs = plt.subplots(2,2)
axs[0,0].set_title('Rnica')
axs[0,0].imshow(np.uint8(I_mandril-I_lena),"gray")
axs[0,0].axis("off")
axs[0,1].set_title('Suma')
axs[0,1].imshow(np.uint8(I_mandril+I_lena),"gray")
axs[0,1].axis("off")
axs[1,1].set_title('Mnoenie')
axs[1,1].imshow(np.uint16(I_mandril*I_lena),"gray")
axs[1,1].axis("off")
axs[1,0].set_title('absdiff')
axs[1,0].imshow(cv2.absdiff(I_mandril,I_lena),"gray")
axs[1,0].axis("off")
plt.show()

lin_comb = np.uint8(0.8 * I_lena + 0.2 * I_mandril)
plt.imshow(lin_comb,"gray")
plt.title("Kombinacja liniowa")
plt.axis("off")
plt.show()
```

## Zadanie 1.6

```python
def hist(img):
    h=np.zeros((256,1), np.float32)
    height, width = img.shape[:2]
    for y in range(height):
        for x in range(width):
            h[int(img[x,y]),0] += 1
    return h

hist_lena = hist(I_lena)
plt.bar(np.arange(256),hist_lena.reshape(256))
plt.title("Histogram obrazu Lena")

hist_lena_cv2 = cv2.calcHist([I_lena],[0],None,[256],[0,256])
print(sum(abs(hist_lena-hist_lena_cv2)))
```

## Zadanie 1.7

```python
I_lena = cv2.imread("lena.png")
I_lena = cv2.cvtColor(I_lena,cv2.COLOR_BGR2RGB)
I_lena_gauss = cv2.GaussianBlur(I_lena,(5,5),1.5)

plt.imshow(I_lena)
plt.title("Lena bez filtracji")
plt.axis("off")
plt.show()

plt.imshow(I_lena_gauss)
plt.title("Filtracja Gaussa")
plt.axis("off")
plt.show()
```

```python
I_lena_sobel = cv2.Sobel(cv2.cvtColor(I_lena,cv2.COLOR_RGB2GRAY),cv2.CV_64F,1,0,ksize=5)

plt.imshow(I_lena_sobel,"gray")
plt.title("Filtracja Sobela")
plt.axis("off")
plt.show()
```

```python
I_lena_lap = cv2.Laplacian(cv2.cvtColor(I_lena,cv2.COLOR_RGB2GRAY),cv2.CV_64F)
plt.imshow(I_lena_lap,"gray")
plt.title("Laplasjan")
plt.axis("off")
plt.show()
```

```python
I_lena_med = cv2.medianBlur(cv2.cvtColor(I_lena,cv2.COLOR_RGB2GRAY),5)
plt.imshow(I_lena_med,"gray")
plt.title("Mediana")
plt.axis("off")
plt.show()
```
